```python
import pandas as pd
import numpy as np
import string


with open('C:\\Users\\amanka5\\Desktop\\temperature.txt') as f:    #open file
    lines1 = f.readlines()                         # read the given data file

    
Data = []

for i in range(len(lines1)):
    Data.append(lines1[i].strip('\n'))
    
```

```python
#print Data
Data
```

```python
#CLEANING THE DATA using data cleaning function
def clean_data(data):                      
   
    date = []
    temperature = []
    data_final = []
    for i in data:
        d = i.strip()                                          
        d = i.translate(str.maketrans('','',string.punctuation))    
        d = [[int(s) for s in line.split(" ")] for line in d.split("\n")]  
        data_final.append(d)             #appending the updated text values to the cleaned data
        
    return data_final  #returing the final cleaned data
```

```python
#creating main df function after cleaning the Data file
df = clean_data(Data)
#print df
df
```

```python
#USING THE SPLIT FUNCTION TO SPLIT DATA INTO TWO PARTS - split1 and splist2
```

```python
def split_data(df):
    split1 = df[0:500]           #making 2 splits
    split2 = df[500:]
    return split1,split2      #returing the two splits
```

```python
#callig function
split1, split2 = split_data(df)
```

```python
#Using Mapper functions - creating dictionary (key, value) pairs from split data , split1 & split2 
```

```python
def mapper(data):
    
    date = []
    temperature = []
    result = []
    data_final = []
    for i in data:
        for j in i:
            for k in range(len(j)):
                if k%2==0:
                    date.append(j[k])
                else:
                    temperature.append(j[k])
    for i in range(len(date)):
        result.append({'date':date[i],'temperature':temperature[i]})
    return result        
   
```

```python
map_split1 = mapper(split1)             #spliting values in split1 
map_split2 = mapper(split2)             #spliting values in split2

mapper2 = map_split1 + map_split2

```

```python
#printing the final_mapper value
print(mapper2)  
```

```python
#Using sorting function to sort output from final mapper
```

```python
def sort(sorted_data):
    new_list = sorted(sorted_data, key=lambda d: d['date'])  
    return new_list
    
```

```python
#sorting the output into variable
y = sort(mapper2)
```

```python
#print y 
y
```

```python
# Split the date into year 
#extracting date & year values
dataset = pd.DataFrame(y)
dataset['date'] = dataset['date'].astype(str)
dataset['Year'] = [y[:4] for y in dataset['date']]
```

```python
#check the dataset
dataset.head() 
```

```python
# Using partition function on key values pairs
```

```python
def partition(d):   #defining partition function  
    dfset1 = d[d['Year'] <= '2015']      #seperating data into two parts, before 2015
    dfset2 = d[d['Year'] > '2015']       #seperating data into two parts, after 2015
    
    return dfset1,dfset2   # returing function after partition
```

```python
partition1, partition2  =  partition(dataset)   

```

```python
#print partition(dataset)  to check partition
partition(dataset)   
```

```python
#Using reducer function

def reducer(output):           #defining the reducer function to group 'year' & aggregting maximum temperature
    reducer_data = output.groupby(['Year'], as_index=False)['temperature'].agg(list) 
    max_list=[]
    for i in range(len(reducer_data)):
        max_list.append(max(reducer_data['temperature'][i]))  
    reducer_data['Max_temperature'] = max_list
    
    
    return reducer_data  #returing the reducer function
```

```python
df_1 = reducer(partition1)        #getting output from the partition function as input for reducer function
```

```python
#print function
print(df_1)
```

```python
df_2 = reducer(partition2)        #getting output from the partition function as input for reducer function
```

```python
#print function
print(df_2)
```

```python
df_1 = df_1.drop('temperature',  axis=1)         #getting desired output as asked in assignment
df_2 = df_2.drop('temperature',  axis=1)
```

```python
#Concatinating to get final result
final_result = pd.concat([df_1, df_2], ignore_index=True)
print(final_result)
```

```python
#using main function on all the functions- 

def main(y):
    df = clean_data(y)                                #on the main df function

    split1, split2 = split_data(df)                  #applying on mapper function
    
    map_split1 = mapper(split1)           
    map_split2 = mapper(split2)            

    mapper2 = map_split1 + map_split2
    y =  sort(mapper2)
   

    partition1, partition2  =  partition(dataset)            #applying on partition function
    
    df_1 = reducer(partition1)                        #applying on reducer function
    df_2 = reducer(partition2)   
    
    df_1 = df_1.drop('temperature',  axis=1)        
    df_2 = df_2.drop('temperature',  axis=1)
    final_result = pd.concat([df_1, df_2], ignore_index=True)  #applying the main function on the final result after using concat function
    
    return final_result     #returing the final result
    
    
```

```python
#calling the main function & printing output
x = main(Data)
print(x)
```
